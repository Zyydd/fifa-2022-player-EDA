```python
print(df.columns)
df
```

```python
skills = ['pace','shooting', 'passing', 'dribbling', 'defending', 'physic']
```

```python
def avg_calc(df_skills) :
        skills = ['pace','shooting', 'passing', 'dribbling', 'defending', 'physic']
        avg_list = []
        for att in skills :
            avg_Skills = df_skills[att].mean() 
            avg_list.append(avg_Skills) 
        avg_df = pd.DataFrame(avg_list).T
        avg_df.columns = skills
        return avg_df

        
                
```

```python
top_10 = mask.sort_values('overall',ascending= False).head(10)
print(mask.sort_values('overall',ascending = False).head(10))
top_10
```

```python
top_10_skills = pd.DataFrame(top_10,columns=skills)
top_10_skills
```

```python
top_ten_avg = avg_calc(top_10_skills)
top_ten_avg
```

```python
plt.figure(figsize=(15,5)) 
ax = sns.pointplot(data=top_ten_avg,color='red')
ax.set_xlabel("Skills")
ax.set_ylabel("skills Value")
ax.set_title("Avg rating for top 10 players",fontsize =20 )


plt.show()
```

# ðŸŽ¯ Key Insights & Observations

```python
top_10_forward = mask[mask['Position']=='Forward'].sort_values('overall',ascending= False).head(10)
top_10_forward
```

```python
top_10_forwards = pd.DataFrame(top_10_forward,columns=skills) 
top_10_forwards
```

```python
avg_top10_forward = avg_calc(top_10_forwards)
avg_top10_forward
```

```python
plt.figure(figsize=(15,5))
ax = sns.pointplot(avg_top10_forward , color='green')
ax.set_xlabel("skills")
ax.set_ylabel("Avg skill value")
ax.set_title("Avg top 10 Forwards skills",fontsize = 22)
plt.show()
```

```python
avg_top10_Midfield = avg_calc(top_10_Midfield)
avg_top10_Midfield
```

```python
plt.figure(figsize=(15,5))
ax = sns.pointplot(avg_top10_Midfield , color='darkblue')
ax.set_xlabel("skills")
ax.set_ylabel("Avg skill value")
ax.set_title("Avg top 10 Midfield skills",fontsize = 22)
plt.show()
```

```python
top_10_Defender = mask[mask['Position']=='Defender'].sort_values('overall',ascending= False).head(10)
top_10_Defender
```

```python
plt.figure(figsize=(15,5))
ax = sns.pointplot(avg_top10_Defender , color='black')
ax.set_xlabel("skills")
ax.set_ylabel("Avg skill value")
ax.set_title("Avg top 10 Defender skills",fontsize = 22)
plt.show()
```

## Comparison between different Positions.

```python
import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='red', label='Forwards')
orange_patch = mpatches.Patch(color='orange', label='Midfielders')
blue_patch = mpatches.Patch(color='blue', label='Defenders')


plt.figure(figsize=(20, 8))

sns.pointplot(data=avg_top10_forward, color='red')
sns.pointplot(data=avg_top10_Midfield, color='orange')
sns.pointplot(data=avg_top10_Defender, color='blue')
plt.title("Forwards vs Midfielders vs Defenders (Top 10)")
plt.xlabel("Skills", fontsize=20)
plt.ylabel("Value", fontsize=20)
plt.grid()

plt.legend(handles=[red_patch, orange_patch, blue_patch]);
```
